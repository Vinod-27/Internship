```python
import selenium 
from selenium import webdriver
import pandas as pd
import csv
import time
import urllib
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException,ElementNotInteractableException,ElementNotVisibleException,TimeoutException
```

```python
driver=webdriver.Chrome()
driver.get("https://economictimes.indiatimes.com/archive/year-2015.cms")
driver.find_element_by_xpath('//a[51][@class="normtxt"]').click()

link=[]
a=[]
Date=[]
Author=[]
Vertical=[]
Headline=[]
Description=[]
for i in WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,'//table[@id="calender"]//tbody//tr//td//a'))):
    link.append(i.get_attribute('href'))


for p in link[:4]:
    
    try:
        time.sleep(5)
        driver.get(p)
        try:
            for j in WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,'//ul[@class="content"]//li//a')))[0:2]:
                a.append(j.get_attribute('href'))
            for i in a:
                driver.get(i)
                try:
                    Date.append(driver.find_element_by_xpath('//div[@class="dtc vam artByline"]').text.split("d:")[1])
                    Author.append(driver.find_element_by_xpath('//span[@class="ag"]').text)
                    Vertical.append(driver.find_element_by_xpath('//a[@class="current"]').text)
                    Headline.append(driver.find_element_by_xpath('//div[1][@class="topPart clearfix tac fixedOnLoad"]').text)
                    Description.append(driver.find_element_by_xpath('//div[@class="artText medium"]').text)
                except TimeoutException as e:
                    print(e)
        except TimeoutException as e:
            print(e)
    except TimeoutException as e:
        print(e)
```

```python
News1={"Dates":Date,"Authors":Author,"Verticals":Vertical,"Headlines":Headline,"Descriptions":Description}
df=pd.DataFrame(News1)
df
# Date.pop(0)
# Author.pop(0)
```
