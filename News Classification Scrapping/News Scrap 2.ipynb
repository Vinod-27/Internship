```python
import selenium
from selenium import webdriver
import pandas as pd
import csv
import time
import urllib
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException,ElementNotInteractableException,ElementNotVisibleException,TimeoutException
```

```python
driver=webdriver.Chrome('chromedriver.exe')
driver.get('https://timesofindia.indiatimes.com/archive/year-2015.cms')
driver.find_element_by_xpath('//a[44][@class="normtxt"]').click()
link_c=[]
link_h=[]
Date=[]
Author=[]
Vertical=[]
Headline=[]
Description=[]

for i in WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,'//table[@id="calender"]/tbody/tr/td/a'))):
    link_c.append(i.get_attribute('href'))
for j in link_c:
    try:
        driver.get(j)
    
        for i in WebDriverWait(driver,5).until(EC.presence_of_all_elements_located((By.XPATH,'//td[@align="left"]/span/a'))):
            link_h.append(i.get_attribute('href'))

        for j in link_h:
            try:
                driver.get(j)
                try:
                    Date.append(driver.find_element_by_xpath('//div[@data-plugin="dateformat"]').text.split("d:")[1])
                except NoSuchElementException as e:
                    Date.append("-")
                try:
                    Author.append(driver.find_element_by_xpath('//div[@itemprop="author"]').text)
                except NoSuchElementException as e:
                    Author.append("-")
                try:
                    Vertical.append(driver.find_element_by_xpath('//a[@pg="Breadcrumb-pos1#Entertainment"]').text)
                except NoSuchElementException as e:
                    Vertical.append("-")
                try:
                    Headline.append(driver.find_element_by_xpath('//h1[@data-plugin="arttitle"]').text)
                except NoSuchElementException as e:
                    Headline.append("-")
                try:
                    Description.append(driver.find_element_by_xpath('//div[@class="article_content clearfix"]').text)
                except NoSuchElementException as e:
                    Description.append("-")
            except TimeoutException as e:
                print(e)
                time.sleep(2)
            
    except TimeoutException as e:
        print(e)
            
```

```python
News={'Dates':Date,'Authors':Author,'Verticals':Vertical,'Headlines':Headline,'Descriptions':Description}
df=pd.DataFrame(News)
df
```

```python
filen="News2.xlsx"
df.to_excel(filen)
```
