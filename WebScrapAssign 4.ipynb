```python
import selenium
from selenium import webdriver
import pandas as pd
import csv
import re
import time
import urllib
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException,ElementNotInteractableException,ElementNotVisibleException,TimeoutException
```

```python
def scrap_you(path,web):
    driver=webdriver.Chrome(path)
    driver.get(web)
    values={
        "Rank":[],
        "Name":[],
        "Artist":[],
        "Upload Date":[],
        "Views":[]
    }
    for i in driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"][1]/tbody/tr'):
        values['Rank'].append(i.find_element_by_xpath('td[1]').text)
        values['Name'].append(re.sub("\[.*?\]","",i.find_element_by_xpath('td[2]').text.replace('"',"")))
        values['Artist'].append(i.find_element_by_xpath('td[3]').text)
        values['Views'].append(i.find_element_by_xpath('td[4]').text)
        values['Upload Date'].append(i.find_element_by_xpath('td[5]').text)
    return values

df=scrap_you(r"C:\Users\veenu\Downloads\chromedriver_win32\chromedriver.exe","https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
pd.DataFrame(df)
```

```python
def scrap_bcc(path,web):
    driver=webdriver.Chrome(path)
    driver.get(web)
    driver.find_element_by_xpath('//div[@class="navigation__drop-down drop-down drop-down--reveal-on-hover"]').click()
    driver.find_element_by_xpath('//a[@class="navigation__link navigation__link--in-drop-down"]').click()
    time.sleep(2)
    values={
        'Title':[],
        'Series':[],
        'Place':[],
        'Date':[],
        'Time':[]
    }
    for t,s,p,d,ti in zip(driver.find_elements_by_xpath('//strong[@class="fixture__name fixture__name--with-margin"]'),
                         driver.find_elements_by_xpath('//div[@class="fixture__teams"]'),
                         driver.find_elements_by_xpath('//div[@class="fixture__description u-unskewed-text"]/p/span'),
                         driver.find_elements_by_xpath('//div[@class="fixture__full-date"]'),
                          driver.find_elements_by_xpath('//span[@class="fixture__time"]')):
        values['Title'].append(t.text)
        values['Series'].append(s.text.replace('\n'," "))
        values['Place'].append(p.text)
        dates=d.text.split("\n")[:2]
        values['Date'].append('-'.join(dates))
        values['Time'].append(ti.text)
    return values
df=scrap_bcc(r"C:\Users\veenu\Downloads\chromedriver_win32\chromedriver.exe","https://www.bcci.tv/")
pd.DataFrame(df)
```

```python
def scrap_selenium_exceptions(path,web):
    driver = webdriver.Chrome(path)
    driver.get(web)
    time.sleep(2)
    driver.find_element_by_id('java_technologies').find_element_by_xpath("li[3]").click()
    driver.find_element_by_xpath("//a[@title='Selenium Exception Handling (Common Exceptions List)']").click()
    values={
        "Name":[],
        "Description":[]}
    for i in driver.find_elements_by_xpath("//table[@class='table table-striped']/tbody/tr"):
        try:
            i.find_element_by_xpath("td/strong").text
        except NoSuchElementException as e:
            values["Name"].append(i.text.split(" ")[0])
            values['Description'].append(' '.join(i.text.split(" ")[1:]))
    return values
data=scrap_selenium_exceptions(r"C:\Users\veenu\Downloads\chromedriver_win32\chromedriver.exe","https://www.guru99.com/")
pd.DataFrame(data)
```

```python
def scrap_stat(path,web):
    driver=webdriver.Chrome(path)
    driver.get(web)
    driver.get(driver.find_element_by_xpath('//div[@class="navbar"]/div[2]/div/a[3]').get_attribute('href'))
    driver.find_element_by_xpath('//div[@style="float:left;background-color:seashell;width:400px;height:800px;"]/ul/li/a').click()
    values={
        'Rank':[],
        'State':[],
        'GSDP20':[],
        'GSDP19':[],
        'Share':[],
        'GDP':[]
    }
    for a,b,c,d,e,f in zip (driver.find_elements_by_xpath('//td[@class="data1"]'),
                           driver.find_elements_by_xpath('//td[2][@class="name"]'),
                           driver.find_elements_by_xpath('//td[3][@class="data"]'),
                           driver.find_elements_by_xpath('//td[4][@class="data sorting_1"]'),
                           driver.find_elements_by_xpath('//td[5][@class="data"]'),
                           driver.find_elements_by_xpath('//td[6][@class="data"]')):
        values['Rank'].append(a.text)
        values['State'].append(b.text)
        values['GSDP20'].append(c.text)
        values['GSDP19'].append(d.text)
        values['Share'].append(e.text)
        values['GDP'].append(f.text)
    return values
df=scrap_stat(r"C:\Users\veenu\Downloads\chromedriver_win32\chromedriver.exe","http://statisticstimes.com/")
pd.DataFrame(df)
```

```python
def scrap_github_data(path,web_link):
    driver = webdriver.Chrome(path)
    driver.get(web_link)
    time.sleep(1)
    driver.get(driver.find_element_by_xpath("//ul[@class='list-style-none mb-3']/li[3]/a").get_attribute("href"))
    values={
        "Repository title":[],
        "Repository description":[],
        "Contributors count":[],
        "Language used":[]}
    for title,desc,count,lang in zip(driver.find_elements_by_xpath("//span[@class='text-normal']"), 
                                     driver.find_elements_by_xpath("//p[@class='col-9 color-text-secondary my-1 pr-4']"), 
                                     driver.find_elements_by_xpath("//div[@class='f6 color-text-secondary mt-2']/a[2]"),
                                     driver.find_elements_by_xpath("//span[@itemprop='programmingLanguage']")):
        values['Repository title'].append(title.text.split()[0])
        values['Repository description'].append(desc.text)
        values['Contributors count'].append(count.text)
        values['Language used'].append(lang.text)
    return values
data=scrap_github_data(r"C:\Users\veenu\Downloads\chromedriver_win32\chromedriver.exe","https://github.com/")
pd.DataFrame(data)
```

```python
def scrap_song(path,web):
    driver=webdriver.Chrome(path)
    driver.get(web)
    driver.get(driver.find_element_by_xpath('//div[@class="header__sticky__child"]/nav/ul/li[3]/a').get_attribute('href'))
    time.sleep(3)
    values={
        'Song':[],
        'Artist':[],
        'Week':[],
        'Peak':[],
        'Chart':[]
    }
    for a,b,c,d,e in zip(driver.find_elements_by_xpath('//span[@class="chart-element__information__song text--truncate color--primary"]'),
                         driver.find_elements_by_xpath('//span[@class="chart-element__information__artist text--truncate color--secondary"]'),
                         driver.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--last"]'),
                         driver.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--peak"]'),
                         driver.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--week"]')):
        values['Song'].append(a.text)
        values['Artist'].append(b.text)
        values['Week'].append(c.text)
        values['Peak'].append(d.text)
        values['Chart'].append(e.text)
    return values
df=scrap_song(r"C:\Users\veenu\Downloads\chromedriver_win32\chromedriver.exe","https://www.billboard.com/")
pd.DataFrame(df)
```

```python
def scrap_job_data_sci_nkry(path,web_link):
    driver = webdriver.Chrome(path)
    driver.get(web_link)
    driver.get(driver.find_element_by_xpath("//a[@title='Search Recruiters']").get_attribute('href'))
    driver.find_element_by_xpath("//input[@class='sugInp']").send_keys("Data Science")
    driver.find_element_by_id("qsbFormBtn").click()
    data=[]
    data.append(["Name","Designation","Company","Location","Skills they hire for"])
    for i in driver.find_elements_by_xpath("//div[@class='recInfo']"):
        row=i.text.split("\n")
        row.pop(4)
        data.append(row[:5])
    return data
data=scrap_job_data_sci_nkry(r"C:\Users\veenu\Downloads\chromedriver_win32\chromedriver.exe","https://www.naukri.com/")
pd.DataFrame(data[1:],columns=data[0])
```

```python
def scrap_imdb(path,web):
    driver=webdriver.Chrome(path)
    driver.get(web)
    values={
        'Name':[],
        'Year':[],
        'Genre':[],
        'Runtime':[],
        'Ratings':[],
        'Votes':[]
    }
    for a,b,c,d,e,f in zip( driver.find_elements_by_xpath('//div[@class="lister-item mode-detail"]/div[2]/h3/a'),
                           driver.find_elements_by_xpath('//div[@class="lister-item mode-detail"]/div[2]/h3/span[2]'),
                           driver.find_elements_by_xpath('//div[@class="lister-item mode-detail"]/div[2]/p/span[5]'),
                           driver.find_elements_by_xpath('//div[@class="lister-item mode-detail"]/div[2]/p/span[3]'),
                            driver.find_elements_by_xpath('//span[@class="ipl-rating-star__rating"]'),
                           driver.find_elements_by_xpath('//span[@name="nv"]')):
        values['Name'].append(a.text)
        values['Year'].append(b.text)
        values['Genre'].append(c.text)
        values['Runtime'].append(d.text)
        values['Ratings'].append(e.text)
        values['Votes'].append(f.text)
    return values
df=scrap_imdb(r"C:\Users\veenu\Downloads\chromedriver_win32\chromedriver.exe","https://www.imdb.com/list/ls095964455/")
pd.DataFrame(df)
```

```python
def scrap_uci(path,web):
    driver=webdriver.Chrome(path)
    driver.get(web)
    driver.find_element_by_xpath('//span[@class="whitetext"]/a/font/b').click()
    data=[]
    for i in driver.find_elements_by_xpath('//table[@border="1"]/tbody/tr'):
        data.append(i.text.split("\n"))
    return data
df=scrap_uci(r"C:\Users\veenu\Downloads\chromedriver_win32\chromedriver.exe","https://archive.ics.uci.edu/")
pd.DataFrame(data[1:],columns=data[0])
    
```
