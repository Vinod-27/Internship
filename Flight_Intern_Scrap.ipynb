```python
30oct
Airline=[]
Date=[]
Source=[]
Destination=[]
Dep_Time=[]
Arrival_Time=[]
Duration=[]
Stops=[]
Price=[]
Date=[]
30 oct
driver=webdriver.Chrome('chromedriver')
driver.get('https://www.yatra.com/')
driver.find_element_by_xpath('//input[@type="submit"]').click()

source=["New Delhi","Mumbai","Bangalore","Chennai","Hyderabad","Goa","Kolkata","Pune","Jaipur","Lucknow"]
destination=["New Delhi","Bangalore","Mumbai","Chennai","Hyderabad","Goa","Kolkata","Pune","Jaipur","Lucknow"]
for i in source:
    search_s=driver.find_element_by_xpath('//input[@id="origin_0"]')
    search_s.clear()
    time.sleep(3)
    search_s.click()
    time.sleep(3)
    search_s.send_keys(i)
    time.sleep(2)
    for j in destination:
        search_d=driver.find_element_by_xpath('//input[@id="destination_0"]')
        search_d.clear()
        time.sleep(3)
        search_d.click()
        time.sleep(3)
        search_d.send_keys(j)
        time.sleep(5)
        if i==j:
            continue
            search_d.clear()
            time.sleep(3)
            search_d.send_keys(j)
            time.sleep(3)
        driver.find_element_by_xpath('//span[@class="radio v-aligm-t"]').click()           
        try:
            WebDriverWait(driver,5).until(EC.element_to_be_clickable((By.XPATH,'//span[@ng-if="!submitted"]'))).click()
        except NoSuchElementException as e:
            print(e)
        time.sleep(3)
        try:
            WebDriverWait(driver,5).until(EC.element_to_be_clickable((By.XPATH,'/html/body/section[2]/section/section[2]/section[1]/div[1]/div/div[1]/div/ul/li[2]/div'))).click()
        except NoSuchElementException as e:
            print(e)
        time.sleep(3)
        last_height = driver.execute_script("return document.body.scrollHeight")
        while True:
        # Scroll down to bottom
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                    # Wait to load page
            time.sleep(2)
                    # Calculate new scroll height and compare with last scroll height
            new_height = driver.execute_script("return document.body.scrollHeight")
            if new_height == last_height:
                break
            last_height = new_height
        driver.find_element_by_tag_name('body').send_keys(Keys.CONTROL + Keys.HOME)
        time.sleep(2)
        for A,S,D,Dep,Ar,St,Du,P in zip(driver.find_elements_by_xpath('//div[@class="fs-13 airline-name no-pad col-8"]/span'),
                                        driver.find_elements_by_xpath('//div[@class="i-b col-4 no-wrap text-right dtime col-3"]/p'),
                                        driver.find_elements_by_xpath('//div[@class="i-b pdd-0 text-left atime col-5"]/p[2]'),
                                        driver.find_elements_by_xpath('//div[@autom="departureTimeLabel"]'),
                                        driver.find_elements_by_xpath('//p[@autom="arrivalTimeLabel"]'),
                                        driver.find_elements_by_xpath('//div[@class=" font-lightgrey fs-10 tipsy i-b fs-10"]'),
                                        driver.find_elements_by_xpath('//p[@autom="durationLabel"]'),
                                        driver.find_elements_by_xpath('//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')):
            Airline.append(A.text)
            Source.append(S.text)
            Destination.append(D.text)
            Dep_Time.append(Dep.text)
            Arrival_Time.append(Ar.text[:5])
            Stops.append(St.text)
            Duration.append(Du.text)
            Price.append(P.text)
        time.sleep(2)
for s in driver.find_elements_by_xpath('//div[@class="scroller full-width flex"]/ul/li[2]/div/p[1]'):
    Date.append(s.text[4:])
    z=len(Airline)
    Date=Date*z
```

```python
d30=pd.DataFrame({})
d30['airline']=Airline
d30['source']=Source
d30['destination']=Destination
d30['deptime']=Dep_Time
d30['arrivtime']=Arrival_Time
d30['stops']=Stops
d30['duration']=Duration
d30['date']=Date
d30['price']=Price
```
